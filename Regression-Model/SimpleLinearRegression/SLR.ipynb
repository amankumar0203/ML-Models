**IMPORT LIBRARIES**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**IMPORT DATASET**

```python
dataset = pd.read_csv("insurance.csv")
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

**SPLITTING THE DATASET INTO TRAINING AND TEST SET**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=3/7, random_state= 0)
```

**TRAINING THE TRAIN SET**

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

**PREDICTING THE TEST RESULT**

```python
y_pred = regressor.predict(x_test)
combined = np.hstack((x_test, y_pred.reshape(-1,1)))
print(combined)
np.set_printoptions(suppress='true', precision=2)
print(combined)
```

**VISUALIZING TRAIN SET RESULT**

```python
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, regressor.predict(x_train),color='blue')
plt.title('Simple Linear Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Premium')
plt.show()
```

**VISUALIZING TEST SET RESULT**

```python
plt.scatter(x_test, y_test, color='red')
plt.plot(x_train, regressor.predict(x_train), color='blue')
plt.title('Simple Linear regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Expected Premium')
plt.show()
```

```python
age = 25
#sklearn always expects it's input to be in 2D array form

#x_train, x_test ... these variables are 1D array still sklearn is able to process it because
#pandas DataFrames and Series, when converted to NumPy arrays using .values (as you've done in your code),
#are automatically treated as 2D arrays by scikit-learn, even if they represent a single feature.
age_reshape = np.array(age).reshape(-1,1)
print(regressor.predict(age_reshape))
```
